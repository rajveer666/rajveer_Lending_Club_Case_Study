```python
import pandas as pd          
#To read csv

import matplotlib.pyplot as plt 
#For plotting

import seaborn as sns       
#Plotting helper library

import numpy as np 
#For mathematical functions

%matplotlib inline 
#To allow plotting inside the notebook

import warnings
warnings.filterwarnings("ignore") 
#To ignore deprecation warnings

loan = pd.read_csv('loan.csv')

pd.set_option('display.max_columns', None) 
#this statement use to show all columns

loan
```

```python
loan=loan.dropna(axis=1,how='all') 
#remove all NaN columns
loan
```

```python
loan.fillna(0,inplace=True) 
#fill all NaN to 0 value
```

```python
loan.isnull().sum()
#print how much NaN values in each column
```

```python
loan.drop(columns=['url','desc','tax_liens','delinq_amnt','application_type','acc_now_delinq','policy_code','initial_list_status','pymnt_plan'],inplace=True)
#drop that columns which contain either zero or NaN values 
loan
```

```python
sns.set(rc={'figure.figsize':(15,10)})
# set the figure size 

plt.bar(loan['int_rate'],loan['home_ownership'],width=10) 
#create bar chart int rate and homeownership with width 10

plt.title("HOME OWNERSHIP WITH INT RATE",fontdict={'fontsize':20, 'fontweight':5,'color':'Green'})
#title HOME OWNERSHIP WITH INT RATE

plt.show()
```

```python
plt.scatter(loan['loan_amnt'],loan['grade'],alpha=0.7)
#plot scatter chart to represent loan amount and grade

plt.title("LOAN WITH GRADES",fontdict={'fontsize':20, 'fontweight':5,'color':'Green'})
#title loan with grades

plt.xlabel("LOAN AMOUNT",fontdict={'fontsize':10, 'fontweight':5,'color':'Red'})
#  x-axis label of the plot

plt.ylabel("GRADE",fontdict={'fontsize':10, 'fontweight':5,'color':'Red'})
#y axis label of the plot
```

```python
plt.scatter(loan['annual_inc'],loan['purpose'],alpha=0.5)
#plot scatter chart to represent annual income and purpose

plt.title("PURPOSE ACCORDING TO INCOME",fontdict={'fontsize':20, 'fontweight':5,'color':'Green'})
#title PURPOSE ACCORDING TO INCOME 

plt.xlabel("ANNUAL INCOME",fontdict={'fontsize':10, 'fontweight':5,'color':'Red'})
# x-axis label

plt.ylabel("PURPOSE",fontdict={'fontsize':10, 'fontweight':5,'color':'Red'})
#y-axis label
```

```python
sns.set(rc={'figure.figsize':(20,20)})
# set the figure size 

plt.scatter(loan['annual_inc'],loan['addr_state'],alpha=0.9)
#plot scatter chart to represent annual_inc and addr_state

plt.title("INCOMES IN DIFFERENT STATES",fontdict={'fontsize':20, 'fontweight':5,'color':'Green'})
#title INCOMES IN DIFFERENT STATES

plt.xlabel("ANNUAL INCOME",fontdict={'fontsize':10, 'fontweight':5,'color':'Red'})
# x-axis label

plt.ylabel("STATES",fontdict={'fontsize':10, 'fontweight':5,'color':'Red'})
# y-axis label
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16,5))
# Making a plot with 2 subplots

loan_amount = loan.loan_amnt.values
funded_amount = loan.funded_amnt.values


sns.distplot(loan_amount, ax=ax[0], color='#000000')
ax[0].set_title("Loan Applied by the Borrower", fontsize=14)
sns.distplot(funded_amount, ax=ax[1])
ax[1].set_title("Amount Funded by the Lender", fontsize=14)
# Distribution plots

# Loan applied for and loan funded have same distribution, meaning most loans were accepted
```

```python
sns.boxplot(y=loan_amount)
loan_amount.mean()
# Most loans were between 10000 and 20000 dollars
```

```python
sns.set(rc={'figure.figsize':(10,10)})
# set the figure size 

plt.title("VERIFICATION STATUS")
#title VERIFICATION STATUS

loan.verification_status.value_counts().plot.pie()
#plot pie chart

plt.show()
```

```python
plt.title("GRADE")
#title GRADE

loan.grade.value_counts().plot.pie()
#plot pie chart on garde column

plt.show()
```
